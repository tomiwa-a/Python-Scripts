```python
numRows = 5

def generate(numRows):
    if numRows == 1: return [[1]]
    if numRows == 2: return [[1], [1,1]]
    pascal = [[1], [1,1]]
    i = 3
    while i <= numRows:
        new_arr = [1]
        solve = pascal[len(pascal)-1]
        for a in range(i-2):
            new_arr.append(solve[a]+ solve[a+1])
        new_arr.append(1)
        pascal.append(new_arr)
        i += 1
    return pascal
        
print(generate(numRows))
```
